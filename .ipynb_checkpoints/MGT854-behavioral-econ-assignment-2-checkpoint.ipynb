### 2. 

Let's compare a few strategies given Shane's objective function.


- A: focus on the best and brightest, moving as quickly as possible
- B: focus on bringing up the rear, making sure everybody learns the same amount
- C: something in between


- In (A), let's say the top 50 students learn X and everybody else learns nothing (way too hard). 50X^.7
- In (B), let's say all 276 students learn Y. S = 276Y^.7
- In (C), lets imagine you lose 50 students, and everyone else learns Z. 226Z^.7

I find that Shane's objective function implies only teaching the brightest 50 kids if they learn 12 times as much as the class would learn if you taught everyone. It suggests only teaching for the top 100 kids if they learn 5 times as much as the default amount.

$$
\begin{equation}
226 Y^{.7} = 276Z^{.7} \\
Y^{.7} = \frac{276}{226}Z^{.7} \\
Y = \frac{276}{226}^{\frac{1}{.7}} Z \\
Y = 1.3 Z
\end{equation}
$$

Since I think it's quite reasonable that the most motivated 50 kids could learn 12 times as much in a course taught at a speed that would challenge them as in a course taught to make everyone learn something, Shane's metric implies abandoning more than 2/3s of the class, which I think the Dean would disapprove of. A much lower discounting rate than .7 might be required.

```python
plt.plot(np.power(276/np.arange(1, 276), 1/.7))
plt.ylim(0,20)
plt.title("Required ratio of learning for equivalent class size to 276 (zoomed)")
plt.show()
```
